```python
## Problem 1 (Mandatory) ##

import requests                #import modules and libraries
from bs4 import BeautifulSoup
import json

def get_scrapped_data(soup):
    output_list = []
    dict1 = {}
    products = soup.find_all("div", attrs={"class":"product"}) # fetch all the product cards

    for product in products:
        title = product.find("a", attrs={"class":"catalog-item-name"}).text # fetch title

        prices = product.find("div", attrs={"class":"catalog-item-price"}) # fetch price
        price = prices.find("span", attrs={"class":"price"}).text

        status = product.find("div", attrs={"class":"price-rating-container"}) # fetch availability of stocks
        stock = status.find("span", attrs={"class":"status"}).text
        if stock == "Out of Stock":                                 # reassign true or false to convert it into json object
            stock = False
        else:
            stock = True

        maftr = product.find("a", attrs={"class":"catalog-item-brand"}).text # fetch manufacturer
        
        dict1["price"]=price
        dict1["title"]=title
        dict1["stock"]=stock
        dict1["maftr"]=maftr           # Initialize dictionary with all the previous fetched value
        
        json_object = json.dumps(dict1, indent=2)       # convert the dict into json format    
        output_list.append(json_object)                 # store all the output into the list
        dic1 ={}                                        # Make it null for next product operation
        
    for output in output_list:
        print(output)                                   # To make it more readable iterate over the list and print the same
    

start_url = "https://www.midsouthshooterssupply.com/dept/reloading/primers?currentpage=1" # starting url
HOME_URL = "https://www.midsouthshooterssupply.com"

count = 0                                                      # Initialize count i.e Number of pages you want to scrap

while count < 2:
    response = requests.get(start_url)                         # reponse for targeted url
    soup = BeautifulSoup(response.content,"html.parser")       # html parsed using BS
    get_scrapped_data(soup)                                    # pass the soup for scrapping
    next_link = soup.find("div", attrs={"class":"pagination"}) # next link fetched from anchor tab
    link_container = []
    for a in next_link.find_all('a', href=True):               # as the anchor tag dont't have specific class associated with next url so we would collect all the url and they are all same
        if a.text: 
            link_container.append(a['href'])
    start_url = HOME_URL + link_container[0]                   # next url is generated by adding Home url and fetched url from next button
    
    count += 1                                                 # increment count 
```
